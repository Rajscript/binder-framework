```python
df1=df.groupby('NTA')['Company Name'].agg(Number_of_business ='count')
df2=df.groupby('NTA')['Total Savings'].agg(Mean_Job_Savings='mean')
df3=df.groupby('NTA')['Job created'].agg(TotalJobsCreated='sum')
df2
```

```python
output=df1.merge(df2,on='NTA').merge(df3,on='NTA')
```
